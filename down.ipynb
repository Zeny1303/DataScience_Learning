```python
from sentence_transformers import SentenceTransformer
import shutil
import os

# Step 1: Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Step 2: Get the local folder where the model is saved
cache_dir = model.model_dir   # <--- works reliably across versions

print("Model is stored at:", cache_dir)

# Step 3: Zip the folder
output_path = shutil.make_archive('all-MiniLM-L6-v2', 'zip', cache_dir)

print(f" Model zipped successfully: {output_path}")

```
